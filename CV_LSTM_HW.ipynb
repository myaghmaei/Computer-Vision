```python
import os
import sys
import shutil
import numpy as np
from keras import layers
from keras import models
from keras import optimizers
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from keras.constraints import maxnorm
from sklearn.model_selection import train_test_split
from keras.layers import LSTM
from keras.utils import to_categorical
from keras.layers import Bidirectional
```

```python
dataset = np.load('action.npy')
```

```python
# calculating frame counts of every video and extracting the maximum count number

temp = dataset[0, 0]
videos_frames = np.array([], dtype="i")
frame_count = 0

for i in range(dataset.shape[0]):
    if temp != dataset[i, 0]:
        videos_frames = np.append(videos_frames, frame_count)
        frame_count = 1
        temp = dataset[i, 0]
    
    else:
        frame_count += 1
        
videos_frames = np.append(videos_frames, frame_count)

max_frame_count = np.amax(videos_frames) # frame count of the video which have the maximum frames
```

```python
# padding section

j = 0
padded_dataset = dataset

for i in range(videos_frames.shape[0]):
    
    needed_pads = max_frame_count - videos_frames[i]
    
    padding_array = np.zeros(dataset.shape[1])
    padding_array[0] = padded_dataset[j, 0]
    padding_array[1] = padded_dataset[j, 1]
    padding_array[2] = padded_dataset[j, 2]
    
#    j = j + videos_frames[i]
    
    for a in range(needed_pads):
        padded_dataset = np.insert(padded_dataset, j, padding_array, 0)
        
    j = j + videos_frames[i]
    j = j + needed_pads
```

```python
# preparing the data (cross validation) for LSTM

x_data = []
y_label = []

for i in range(padded_dataset.shape[0]):
    x_data = np.append(x_data, padded_dataset[i, 3:48])
    
for i in range(0, padded_dataset.shape[0], max_frame_count):
    y_label = np.append(y_label, padded_dataset[i, 2])
    
x_data = x_data.reshape(int(padded_dataset.shape[0] / max_frame_count), max_frame_count, padded_dataset.shape[1] - 3)
y_label = to_categorical(y_label)
```
